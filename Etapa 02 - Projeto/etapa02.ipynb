```python
from sqlalchemy import create_engine
from sqlalchemy.orm import declarative_base
from sqlalchemy import Column, Integer, String, Date
import pandas as pd
```

```python
dados_csv = pd.read_csv("banco_de_dados.csv",
                       sep = ";",
                       index_col = "Date")

dados_csv
```

```python
#Mudar coluna do database

dados_csv.columns = ["mrve",
                         "trpl",
                         "petr",
                         "viia",
                         "mglu",
                         "bbdc"]

dados_csv
```

```python
#Conectando ao banco de dados
user = 'root'
senha = 'guerra91'
banco_de_dados = 'dados'
host = 'localhost' #Caso seja em nuvem só colocar o link dela
porta = 3306

conexao = create_engine(url = f'mysql+pymysql://{user}:{senha}@{host}:{porta}/{banco_de_dados}')

```

```python
#Criando a tabela
ativos = ["MRVE3.SA", "TRPL4.SA", "PETR4.SA",
           "VIIA3.SA", "MGLU3.SA", "BBDC4.SA"]
Banco_de_dados = declarative_base()

class User(Banco_de_dados):
    
    __tablename__ = 'Cotacoes'
    
    id = Column(Integer, primary_key = True, autoincrement = True)
    Date = Column(String(50))
    mrve = Column(String(50))
    trpl = Column(String(50))
    petr = Column(String(50))
    viia = Column(String(50))
    mglu = Column(String(50))
    bbdc = Column(String(50))
    
Banco_de_dados.metadata.create_all(conexao)
```

```python
#Colocando dados na tabela utilizando o pandas

nome_tabela = 'Cotacoes'

dados_csv.to_sql(nome_tabela, conexao, index = False, if_exists = 'append')
```
