```python
import yfinance as yf
import pandas as pd
```

```python
ativos = ["MRVE3.SA", "TRPL4.SA", "PETR4.SA",
           "VIIA3.SA", "MGLU3.SA", "BBDC4.SA"]

df_ativos = yf.download(ativos)["Adj Close"]

df_ativos = df_ativos.dropna()

df_ativos.columns = ativos

df_ativos
```

```python
for ativo in ativos:

    df_ativos[ativo] = df_ativos[ativo].map(lambda x: "{:.2f}".format(x))
    
df_ativos
```

```python

for ativo in ativos:
    
    df_ativos[ativo] = df_ativos[ativo].astype(str)
    df_ativos[ativo] = df_ativos[ativo].map(lambda x: "R$" + x)
    
    
print(df_ativos)
```

```python
df_ativos.to_csv("banco_de_dados.csv", sep = ";", index=True, line_terminator='\n', encoding='utf-8')
```
